```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
msk = np.random.rand(len(credit)) < 0.8 #Esto selecciona aleatoriamente el 80% de las observaciones y la amacena en msk
train = credit[msk] #aca se indica que el set de entrenamiento esta conformado por el 80% de las observaciones
test = credit[~msk] #aca se especifica que el set de prueba esta conformado por el 20% de la data restante.
```

```python
plt.scatter(train.LIMIT_BAL, train.AGE,  color='blue')
plt.xlabel("LIMIT_BAL")
plt.ylabel("AGE")
plt.show()
```

```python
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['LIMIT_BAL']])
train_y = np.asanyarray(train[['AGE']])
regr.fit(train_x, train_y)
# Obtener coeficientes de resultados, estos datos son informativos.
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.LIMIT_BAL, train.AGE,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("LIMIT_BAL")
plt.ylabel("AGE")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['LIMIT_BAL']])
test_y = np.asanyarray(test[['AGE']])
test_y_ = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```

```python
credit = pd.DataFrame({'Actual': test_y.flatten(), 'Predicted': test_y_.flatten()})
credit.head(10)
```
